#### 书上实例

```python
import numpy as np
import pandas as pd
```

```python
# loaddataset
def load_dataset(file_path):
    data = pd.read_csv(file_path, sep='\t', header=None).to_numpy()
    return data[:, :-1], data[:, -1]
```

```python
file = 'datasets/testSet.txt'
X, y = load_dataset(file)
```

```python
X,y
```

```python
# file
def get_data(file):
    X=[]
    y=[]
    with open(file) as f:
        for line in f.readlines():
            line_arr=line.strip().split('\t')
            X.append([line_arr[0],line_arr[1]])
            y.append(line_arr[2])
    return X,y
```

```python
get_data(file)[1]
```

```python
def sigmoid(z):
    """
    sigmoid function
    """
    return 1.0/(1+np.exp(-z))
```

```python
# x0=1.0
X=np.column_stack((np.ones(len(X)),X))
X[:10]
```

```python
def grad_ascent(X, y, alpha=0.001, max_cycles=500):
    X = np.mat(X) # 100 *3
    y = np.mat(y).T # 100*1
    m, n = X.shape
    weights = np.ones((n, 1)) # 3*1
    for i in range(max_cycles):
        # w^T * x +b
        h = sigmoid(X * weights)
        error = y - h
        # 3*100 100*1
        weights += alpha * X.T * error
    return weights
```

```python
weights=grad_ascent(X,y)
weights
```

```python
from matplotlib import pyplot as plt


%matplotlib inline
def plot_best_fit(X, y, weights):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.scatter(X[y == 1, 1], X[y == 1, 2], s=30, c='red', marker='s')
    ax.scatter(X[y == 0, 1], X[y == 0, 2], s=30, c='green')
    x0 = np.arange(-3.0, 3.0, 0.1)
    y0 = (-weights[0] - weights[1] * x0) / weights[2]
    ax.plot(x0, y0)
    plt.xlabel('X1')
    plt.ylabel('X2')
```

```python
plot_best_fit(X,y,weights)
```

```python
# 随机梯度上升 一次仅采用一个样本点更新回归系数
def stoc_grad_ascent0(X, y, alpha=0.01, max_cycles=500):
    # X = np.mat(X) # 100 *3
    # y = np.mat(y).T # 100*1
    m, n = X.shape
    weights = np.ones(n) # 3
    for i in range(m):
        # w^T * x +b
        h = sigmoid(np.sum(X[i] * weights))
        error = y[i] - h
        # 3 
        weights += alpha * X[i] * error
    return weights
```

```python
stoc_grad_ascent0(X,y)
```

```python
plot_best_fit(X,y,stoc_grad_ascent0(X,y,0.004))
```

```python
# 测算迭代时间图
```

```python
# 根据迭代收敛看出，可调整alpha为随机值
def stoc_grad_ascent1(X, y, num_iter=150):
    m, n = X.shape
    weights = np.ones(n)
    for j in range(num_iter):
        for i in range(m):
            alpha = 4 / (1.0 + j + i) + 0.01
            # [`low`, `high`)
            rand_index = np.random.randint(0, m)
            h = sigmoid(np.sum(X[rand_index] * weights))
            error = y[rand_index] - h
            weights += alpha * X[rand_index] * error
    return weights
```

```python
plot_best_fit(X,y,stoc_grad_ascent1(X,y))
```

```python
# iter 500
plot_best_fit(X,y,stoc_grad_ascent1(X,y,num_iter=500))
```

```python
train_data=pd.read_csv("datasets/horseColicTraining.txt",sep='\t',header=None)
train_data.head()
```

```python
test_data=pd.read_csv("datasets/horseColicTest.txt",sep='\t',header=None)
test_data.head()
```

```python
def classify_vector(z,weights):
    prob=sigmoid(np.sum(z*weights))
    return 1 if prob>0.5 else 0
```

```python
# 合并训练集和测试集
train_data=train_data.to_numpy()
test_data=test_data.to_numpy()
```

```python
X_train,y_train=train_data[:,:-1],train_data[:,-1]
X_train,y_train
```

```python
X_test,y_test=test_data[:,:-1],test_data[:,-1]
X_test,y_test
```

```python
train_weights=stoc_grad_ascent1(X_train,y_train,num_iter=500)
train_weights
```

```python
y_pred=np.array([classify_vector(x,train_weights) for x in X_test])
y_pred
```

```python
y_test==y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
def fetch_housing_data(housing_url=HOUSING_URL, data_path="datasets/"):
    # download dataset
    tgz_path = os.path.join(data_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    # tgz
    with tarfile.open(tgz_path) as tar:
        tar.extractall(data_path)
```

```python
def split_train_test_use(data, test_ratio=0.2, random_state=42):
    import numpy as np
    data_length = len(data)
    # set seed
    np.random.seed(random_state)
    # permutation 返回洗牌后的矩阵 shuffle 只是洗牌
    suffled_indices = np.random.permutation(data_length)
    test_set_size = int(data_length * test_ratio)
    return data.iloc[suffled_indices[test_set_size:]], data.iloc[
        suffled_indices[:test_set_size]]
```

```python
X, y = split_train_test_use(housing)
X.shape, y.shape
```

```python
# 防止更新数据集改变性质，采取hash 唯一id方式
import hashlib


def test_set_check(identifier, test_ratio, myhash):
    import numpy as np
    return myhash(np.int64(identifier)).digest()[-1] < 256 * test_ratio


def split_train_test_by_id(data, test_ratio, id_column, myhash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, myhash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
# 取行索引作为标识列,确保数据增删只能在末尾
housing_with_id = housing.reset_index()
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
train_set,test_set
```

```python
# 取唯一列作为标识
# 经纬度
housing_with_id["id"]=housing["longitude"]*1000+housing["latitude"]
train_set,test_set=split_train_test_by_id(housing_with_id,0.2,"id")
train_set,test_set
```

```python
# 限制收入中位数类别
housing["income_cat"] = np.ceil(housing["median_income"] / 1.5)
# 大于5 均为类别5
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace=True)
```

```python
# 分层抽样
def get_split_by_income(X, y, test_size=0.2, random_state=42):
    from sklearn.model_selection import StratifiedShuffleSplit
    split = StratifiedShuffleSplit(n_splits=1,
                                   test_size=test_size,
                                   train_size=None,
                                   random_state=random_state)
    for train_index, test_index in split.split(X, y):
        return X.loc[train_index], X.loc[test_index]
```
